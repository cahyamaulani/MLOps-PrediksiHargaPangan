# Feature Engineering - Prediksi Harga Pangan

Tahap ini bertujuan membuat fitur time-series dari dataset harga pangan.

Dataset:
- Provinsi: Jawa Timur
- Komoditas: Beras

Feature yang dibuat:
- lag_1 (harga kemarin)
- lag_7 (harga minggu lalu)
- rolling_mean_7 (rata-rata harga 7 hari)

```python
import pandas as pd
```

# Load dataset

```python
df = pd.read_csv("../data/processed/harga_clean.csv")

df["Tanggal"] = pd.to_datetime(df["Tanggal"])

df.head()
```

```python
df.info()
```

# Membuat lag fitur

```python
df["lag_1"] = df["Nilai"].shift(1)

df["lag_7"] = df["Nilai"].shift(7)

df["rolling_mean_7"] = df["Nilai"].rolling(7).mean()

df["rolling_std_7"] = df["Nilai"].rolling(7).std()
```

# Hapus Missing Value

```python
df = df.dropna()

df.head()
```
